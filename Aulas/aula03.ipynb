### Agrupamento

```python
data = pd.read_csv( 'datasets/kc_house_data.csv' )
```

```python
data.head()
```

```python
data['bedrooms'] == 0
```

```python
data[data['bedrooms']==0]
```

```python
print(data[data['bedrooms']==0].shape)
print(data[data['bedrooms']==1].shape)
print(data[data['bedrooms']==2].shape)
print(data[data['bedrooms']==3].shape)
print(data[data['bedrooms']==4].shape)
```

```python
data[['id','bedrooms']]
```

```python
data[['id','bedrooms']].groupby('bedrooms').size()
```

```python
df_grouped = data[['id','bedrooms']].groupby('bedrooms') #groupby (separa, aplica,combina)
```

```python
for bedrooms, frame in df_grouped:
    print('numero de quartos: {}'.format(bedrooms))
    print(frame.shape,end='\n\n')
```

```python
for bedrooms, frame in df_grouped:
    print('numero de quartos: {}'.format(bedrooms))
    print(frame.head(),end='\n\n')
```
